Learning Vincent

We downloaded DetailedStateInfoAP-CS-A-2006-2013-with-PercentBlackAndHIspanicByState-fixed.xlsx [from this website](http://home.cc.gatech.edu/ice-gt/556).

```python
#you have to import pandas use the pandas library's datatypes and methods
#we aliased them to pd because it's hip, and shorter to type than pandas.

import pandas as pd
```

```python
# if you open this file in excel, it's got 3 sheets titled "Sheet1", "Sheet2", and "Sheet4"
# we specifically imported the second sheet of the excel file, titled "Sheet2"
# we used the convenient pandas.read_excel method, which returns a pandas.DataFrame datatype
# we assigned this output to a variable called "df"

df = pd.read_excel("DetailedStateInfoAP-CS-A-2006-2013-with-PercentBlackAndHIspanicByState-fixed.xlsx",sheetname="Sheet2")
```

```python
# I always like to look at the first couple lines of my DataFrame with the pandas.DataFrame.head() method
# IPython notebooks make them look all sexy and HTML-y
# so I can check that it imported properly in an aesthetically pleasing way

df.head()
```

```python
# next I like to use the pandas.DataFrame.describe() method to get some summary stats about my columns
df.describe()
```

```python
# Now we want to actually import vincent so we can access its classes and methods. 
# We aliased the package name to "v"

import vincent as v

# You have to use the vincent.core.initialize_notebook() method to make graphs display inline in IPython notebook
# similar concept as the '%pylab inline' magic of GADSDC times

v.core.initialize_notebook()
```

```python
# the vincent.Bar class makes bar graphs
# give it a list of values or something that can be coerced to a list of values

bar = v.Bar(df['# schools'])
bar.display()
```

```python
# the vincent.Scatter class makes scatter plots
# it expects the data (X and Y values) as lists mapped to a python dictionary
# tell vincent which one should be on the X axis with the `itr_idx` argument

scatter = v.Scatter({"index":df['# schools'],"index2": df["Total #"]}, iter_idx="index")
scatter.display()
```

```python
# this is how we figured out what to pass vincent for the scatter plot data
# it's the code used to generate the  data used for the Vincent documentation
# all that is listed under "Data" on the vincent QuickStart page
# http://vincent.readthedocs.org/en/latest/quickstart.html#data
# detective work is key

import random
cat_1 = ['y1', 'y2', 'y3', 'y4']
index_1 = range(0, 21, 1)
multi_iter1 = {'index': index_1}
for cat in cat_1:
    multi_iter1[cat] = [random.randint(10, 100) for x in index_1]
```

```python
# the cool thing about vincent is that it can export your static python graph into a Vega JSON object
# Vega is a JSON grammar that the Vega JS library can use to make interactive graphs
# http://trifacta.github.io/vega/editor/index.html?spec=bar
# ^ maybe next time we'll do something like that with the Vega JSON output we exported below

import json
with open("file.json", "w") as f:
    json.dump(scatter.to_json(), f)
```
